# List

## List Basics

```python
# array - homogeneous structure
# list - heterogeneous structure
lst = list()
print(type(lst))
lst = [10,20,30,40,60]
print(len(lst))
print(lst[0])
print(lst[-2])
```

```python
marks = []
marks.append(78)
print(marks)
```

```python
marks.append(80) # adds an element at the end of the list
print(marks)
marks.insert(1,'AI') # adds an element at the specified position
print(marks)
# marks.pop() # removes an element from the end of the list
marks.pop(1) # removes an element from the specified position
print(marks)
marks.remove(80) # removes an element from the list if it exists
print(marks.index(78))
```

```python
# searching
target = 80
if target in marks:
  marks.remove(target)
print(marks)
```

## List comprehension method

```python
squared = [i**2+3 for i in range(1,11) if i % 2 == 1]
print(squared)
```

```python
lst = [10,20,30,40,60,100,200]
#Iteration: index-based
for i in range(len(lst)):
  print(lst[i])

# Iteration: element-based
for i in lst:
  print(i)
```

## List Slicing

```python
squared = [i**2 for i in range(1,11)]
print(squared)
first = []
for i in range(0,5):
  first.append(squared[i])
print(first)
odd_squared = squared[::2]
print(odd_squared)
even_squared = squared[1::2]
print(even_squared)
ind = len(squared)//2
middle = squared[ind:ind+3]
print(middle)
```

```python
# List reverse
squared = squared[7:2:-1]
print(squared)
```
