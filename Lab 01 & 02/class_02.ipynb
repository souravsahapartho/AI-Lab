## List Comprehension Method


```python
squared = [i**2 for i in range(10) if i % 2 == 0]
print(squared)
```

```python
marks = []
marks.append(10)
print(marks)
```

```python
squared = [i**2 for i in range(10) if i % 2 == 0]
print(squared)
first = []
for i in range(0, 3):
    first.append(squared[i])
print(first)
```

### List Slicing


```python
squared = [i**2 for i in range(100)]
print(squared)
first = squared[2:5:2]
print(first)
odd = squared[:5:3]
print(odd)
```

```python
##List Reverse

squared = [i**2 for i in range(10)]
squared = squared[7:2:-1]
print(squared)
```

### List Copy


```python
lst1 = [10, 20, 30, "AI", "Vacation"]
print(lst1)
lst2 = lst1
print(lst2)
lst2.append(200)
print(lst2)
print(lst1)
```

```python
lst3 = lst1.copy()
lst3.pop()
print(lst1)
print(lst2)
print(lst3)
```

```python
# mutable - list, string, dictionary
# imutable - int, float, tuple

a = "AI Lav at very"
print(a)
print(a + "b")
a = a.replace("Lav", "Lab")
print(a)
print(a.index("b"))
```

```python
lst1.append([20, 30, 40])
print(lst1)
lst2 = lst1.copy()  ###shallow copy
print(lst2)
# lst2.remove(20)
# print(lst1,lst2)


print(lst2[-1][0])
lst2[-1].pop()
print(lst2)
print(lst1)
```

```python
## DEEP COPY

import copy

lst3 = copy.deepcopy(lst1)
print(lst3)
lst2[-1].pop()
lst3[-1].pop(1)
print(lst1)
print(lst2)
print(lst3)
```

## Numpy Library


```python
lst1 = [1, 2, 3]
lst2 = [4, 5, 6]
print(lst1 + lst2)
print(lst1 * 3)
```

```python
import numpy as np

arr1 = np.array(lst1)
print(type(arr1))
arr2 = np.array(lst2)
print(type(arr2))
print(arr1 + arr2)
print(arr1 * 3)
print(arr1.shape)
```

```python
arr = np.ones((3,), "int64")  # int64 comma remove kore
print(arr)
arr = np.ones((3, 4), "int64")  # int64 comma remove kore
print(arr)
arr = np.zeros((3, 3), "int64")
print(arr)
```
