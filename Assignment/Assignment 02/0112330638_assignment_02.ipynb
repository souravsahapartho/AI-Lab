## KNN Classification

```python
import numpy as np
import csv
```

```python
train_data = []
validation_data = []
testing_data = []

with open("iris.csv", "r") as file:

    csv_reader = csv.reader(file)
    heading = next(csv_reader)

    for sample_row in csv_reader:

        random_value = np.random.random()

        if random_value >= 0 and random_value <= 0.7:
            train_data.append(sample_row)

        elif random_value > 0.7 and random_value <= 0.85:
            validation_data.append(sample_row)

        else:
            testing_data.append(sample_row)

print("Training Size :", len(train_data))
print("Validation Size :", len(validation_data))
print("Testing Size :", len(testing_data))
```

```python
def calculate_distance(point1, point2):

    feature_a = np.array(point1[1:-1], dtype=float)
    feature_b = np.array(point2[1:-1], dtype=float)

    return np.sqrt(np.sum((feature_a - feature_b) ** 2))
```

```python
def perform_knn(training_set, checking_set, neighbor_count):

    matched = 0
    total_items = len(checking_set)

    for current_item in checking_set:

        distance_store = []

        for trained_item in training_set:

            euclidean_value = calculate_distance(current_item, trained_item)

            distance_store.append((trained_item, euclidean_value))

        distance_store.sort(key=lambda item: item[1])

        nearest_neighbors = distance_store[:neighbor_count]

        predicted_classes = [data[0][-1] for data in nearest_neighbors]

        final_prediction = max(set(predicted_classes), key=predicted_classes.count)

        if final_prediction == current_item[-1]:
            matched += 1

    accuracy_result = (matched / total_items) * 100

    return accuracy_result
```

```python
neighbor_values = [1, 3, 5, 10, 15]

print("Validation Accuracy Results:\n")

for value in neighbor_values:

    validation_result = perform_knn(train_data, validation_data, value)

    print(f"K = {value:2d} --> Accuracy = {validation_result:.2f}%")
```

```python
final_test_accuracy = perform_knn(train_data, testing_data, 5)

print(f"\nFinal Test Accuracy with K=5 : {final_test_accuracy:.2f}%")
```

# KNN Regression


```python
training_samples = []
validation_samples = []
testing_samples = []

with open("diabetes.csv", "r") as file:

    csv_reader = csv.reader(file)
    heading = next(csv_reader)

    for data_row in csv_reader:

        random_number = np.random.random()

        if random_number >= 0 and random_number <= 0.7:
            training_samples.append(data_row)

        elif random_number > 0.7 and random_number <= 0.85:
            validation_samples.append(data_row)

        else:
            testing_samples.append(data_row)

print("Training Size :", len(training_samples))
print("Validation Size :", len(validation_samples))
print("Testing Size :", len(testing_samples))
```

```python
def distance_measure(sample1, sample2):

    first_point = np.array(sample1[:-1], dtype=float)
    second_point = np.array(sample2[:-1], dtype=float)

    return np.sqrt(np.sum((first_point - second_point) ** 2))
```

```python
def regression_knn(train_group, check_group, k_value):

    total_error = 0

    for validation_item in check_group:

        all_distances = []

        for trained_item in train_group:

            calculated_distance = distance_measure(validation_item, trained_item)

            all_distances.append((trained_item, calculated_distance))

        all_distances.sort(key=lambda item: item[1])

        nearest_items = all_distances[:k_value]

        predicted_outputs = [float(data[0][-1]) for data in nearest_items]

        average_prediction = np.mean(predicted_outputs)

        actual_output = float(validation_item[-1])

        total_error += (actual_output - average_prediction) ** 2

    mse_result = total_error / len(check_group)

    return mse_result
```

```python
k_numbers = [1, 3, 5, 10, 15]

print("Validation Mean Squared Error:\n")

for number in k_numbers:

    validation_mse = regression_knn(training_samples, validation_samples, number)

    print(f"K = {number:2d} --> MSE = {validation_mse:.4f}")
```

```python
test_mse_result = regression_knn(training_samples, testing_samples, 5)

print(f"\nTest MSE with K=5 : {test_mse_result:.4f}")
```
