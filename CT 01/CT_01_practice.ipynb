```python
# Problem - 15

courses={
    'Alice':[85,90,78],
    'Bob': [70,75,80],
    'Charlie': [92,88,95]
}

import numpy as np

avg_dict={}

for name,marks in courses.items():
    avg=np.mean(marks)
    avg_dict[name]=avg

print(avg_dict)


# greater than 85

great85=[name for name, avg in avg_dict.items() if avg>85]
print(great85)

# top students and their score

top_student=max(avg_dict, key=avg_dict.get)
top_tuple=(top_student,avg_dict[top_student])

print(top_tuple)
```
